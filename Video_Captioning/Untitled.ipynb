```python
import os
import numpy as np
import json
import string
from random import randint
import math
```

```python
path = "../../data/MLDS_hw2_data/"
traindir = "training_data/feat/"
```

```python
trainjsonfile = open(path+"training_label.json","r")

trainjson = json.load(trainjsonfile)
```

```python
words = []
for x in trainjson:
    for y in x['caption']:
        y = ''.join(c for c in y if c not in string.punctuation)
        ss = str.split(y," ")
        for z in ss:
            words.append(z.lower())
```

```python
wordsFreq = {}
for x in words:
    if x not in wordsFreq:
        wordsFreq[x] = 1
    else:
        wordsFreq[x] = wordsFreq[x] + 1
            
encodeWords = {}
counter = 3
freq = 0
        
for key,value in wordsFreq.items():
    if value>freq:
        encodeWords[key] = counter
        counter = counter + 1
encodeWords["<PAD>"] = 0      
encodeWords["<BOS>"] = 1
encodeWords["<EOS>"] = 2
encodeWords["something"] = 3
print(len(encodeWords))

decodeWords = {}
for key, value in encodeWords.items():
    decodeWords[value] = key
print(len(decodeWords))
```

```python
f = open("raw_text.txt","w")
for x in trainjson:
    
    name = x["id"]
    for y in x["caption"]:

        y = ''.join(c for c in y if c not in string.punctuation)

        temp = []
        for x in str.split(y," "):
            w = x.lower()
            if wordsFreq[w]>freq:
                temp.append(encodeWords[w])
            else:
                temp.append(encodeWords["something"])

        temp2 = "<BOS> "
        for x in temp:
            temp2 = temp2+decodeWords[x]+" "
        temp2 += "<EOS>"
        f.write(temp2+"\n")
```

```python
import fasttext
```

```python
model = fasttext.skipgram('raw_text.txt', 'model', dim=256)
```

```python
print(len(model.words))
print(model["<EOS>"])
```

```python
bbb = np.zeros([2130,256],'float32')
```

```python
encodeW = {}
decodeW = {}
encodeW["<PAD>"] = 0
decodeW[0] = "<PAD>"
counter = 1
for x in model.words:
    encodeW[x] = counter
    decodeW[counter] = x
    counter = counter + 1
print(len(decodeW))
print(len(encodeW))
```

```python
np.save('encodeW.npy',encodeW)
np.save('decodeW.npy',decodeW)
```

```python
bbb[0] = [0]*256
for i in range(1,len(encodeW)):
    x = decodeW[i]
    v= np.array(model[x])
    bbb[encodeW[x]] = v
```
